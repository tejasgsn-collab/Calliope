This section works to convert a dictionary of a certain structure into a dict of a much simpler,flatter structure.

```python
v = {'a': ('int', '10'), 'b': ('int', '100'), 'c': ('float', '10.0')}
```

```python
import ast
```

```python
a = ast.literal_eval(v['a'][1])
print(a, type(a))
```

```python
def convert(variables):
    new = {}
    for i in variables:
        value = ast.literal_eval(variables[i][1])
        new[i] = value
    return new
```

```python
convert(v)
```
